```python
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.RandomState(10)  # deterministic random data
a = np.hstack((rng.normal(size=1000), rng.normal(loc=5, scale=2, size=1000)))
plt.hist(a, bins=30)  # arguments are passed to np.histogram
plt.title("Histogram with manual bins selection")
plt.show()
```

```python
plt.hist(a, bins='sturges')  # arguments are passed to np.histogram
plt.title("Histogram with 'sturges' bins")
plt.show()
```

#### Sturges histogram is much better than the manual selection if we compare both the histograms
#### We can see pattern with sturges as it has optimal bins whereas manual selection has too many bins
